```python
training_region = (
    df_ausbildungssuche_1_berufsausbildung
    .groupby(["angebot.bildungsanbieter.adresse.ortStrasse.land.name", "Kldb_code"])
    .size()
    .reset_index(name="count")
)
```

```python
training_region
```

```python
pivot_table = training_region.pivot(
    index="Kldb_code",
    columns="angebot.bildungsanbieter.adresse.ortStrasse.land.name",
    values="count"
).fillna(0)
```

```python
pivot_table
```

```python
top10 = pivot_table.sum(axis=1).sort_values(ascending=False).head(10)

pivot_top10 = pivot_table.loc[top10.index]
```

```python
pivot_top10
```

```python
import matplotlib.pyplot as plt

pivot_top10.plot(
    kind="bar",
    figsize=(12,7)
)

plt.title("Top Training Occupations by Region")
plt.xlabel("Kldb_code")
plt.ylabel("Number of Training Offers")

plt.xticks(rotation=45)
plt.legend(title="Region")

plt.tight_layout()
plt.show()
```

```python
top_regions = df_ausbildungssuche_1_berufsausbildung["angebot.bildungsanbieter.adresse.ortStrasse.land.name"].value_counts().head(5).index

pivot_filtered = pivot_top10[top_regions]

pivot_filtered.plot(kind="bar", figsize=(12,7))

plt.title("Top Training Occupations by Region")
plt.xlabel("KldB_code (5-digit)", fontsize=14, labelpad=16)
plt.ylabel("Count", fontsize=14, labelpad=16)

plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)

plt.legend(title="Region")

plt.tight_layout()
plt.savefig("../pictures/Top Trainings by Region.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    pivot_top10,
    cmap="YlGnBu",
    annot=True,
    fmt=".0f",
    # cbar_kws={"orientation": "horizontal"}
)

plt.title("Training Occupations by Region (Heatmap)", fontsize=16, pad=30)
plt.xlabel("Regions", fontsize=14, labelpad=16)
plt.ylabel("KldB_code (5-digit)", fontsize=14, labelpad=16)

# Tick labels (axis numbers/text)
plt.xticks(fontsize=12)
plt.yticks(rotation=0,fontsize=12)

plt.tight_layout()
plt.savefig("../pictures/top_trainings_in_region_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Finding new records

```python
new_records = df_ausbildungssuche_2_berufsausbildung[~df_ausbildungssuche_2_berufsausbildung["id"].isin(df_ausbildungssuche_1_berufsausbildung["id"])]
```

```python
new_records.shape
```

```python
old_records = df_ausbildungssuche_2_berufsausbildung[df_ausbildungssuche_2_berufsausbildung["id"].isin(df_ausbildungssuche_1_berufsausbildung["id"])]
```

```python
old_records.shape
```

```python
removed_records = df_ausbildungssuche_1_berufsausbildung[~df_ausbildungssuche_1_berufsausbildung["id"].isin(df_ausbildungssuche_2_berufsausbildung["id"])]
```

```python
removed_records.shape
```

```python
freq_old = df_ausbildungssuche_1_berufsausbildung["Kldb_code"].value_counts()
freq_new = df_ausbildungssuche_2_berufsausbildung["Kldb_code"].value_counts()
```

```python
comparison = pd.concat([freq_old, freq_new], axis=1)
comparison.columns = ["old", "new"]
comparison = comparison.fillna(0)
```

```python
comparison["change"] = comparison["new"] - comparison["old"]
```

```python
top_increase = comparison.sort_values("change", ascending=False).head(10)

top_increase["change"].plot(kind="barh", figsize=(10,6))

plt.title("Increase in Training Offers")
plt.xlabel("Change in Number of Trainings", fontsize=14, labelpad=16)
plt.ylabel("KldB_code (5-digit)", fontsize=14, labelpad=16)

plt.savefig("../pictures/top_increased_trainings.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
top_decrease = comparison.sort_values("change").head(10)

top_decrease["change"].plot(kind="barh", figsize=(10,6))

plt.title("Decrease in Training Offers")
plt.xlabel("Change in Number of Trainings", fontsize=14, labelpad=16)
plt.ylabel("KldB_code (5-digit)", fontsize=14, labelpad=16)

plt.savefig("../pictures/top_decreased_trainings.png", dpi=300, bbox_inches="tight")
plt.show()
```
