```python
region = (
    df_bewerberboerse_1
    .groupby(["lokation.region", "Kldb_code"])
    .size()
    .reset_index(name="count")
)
```

```python
region
```

```python
pivot_table = region.pivot(
    index="Kldb_code",
    columns="lokation.region",
    values="count"
).fillna(0)
```

```python
pivot_table
```

```python
top10 = pivot_table.sum(axis=1).sort_values(ascending=False).head(10)

pivot_top10 = pivot_table.loc[top10.index]
```

```python
pivot_top10
```

```python
import matplotlib.pyplot as plt

pivot_top10.plot(
    kind="bar",
    figsize=(12,7)
)

plt.title("Top applicants by Region")
plt.xlabel("Occupation")
plt.ylabel("Number of applicants")

plt.xticks(rotation=45)
plt.legend(title="Region")

plt.tight_layout()
plt.show()
```

```python
top_regions = df_bewerberboerse_1["lokation.region"].value_counts().head(5).index

pivot_filtered = pivot_top10[top_regions]

pivot_filtered.plot(kind="bar", figsize=(12,7))

plt.title("Top applicants by Region")

plt.xlabel("KldB_code (5-digit)", fontsize=14, labelpad=16)
plt.ylabel("Count", fontsize=14, labelpad=16)

plt.xticks(rotation=0)
plt.legend(title="Region")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,8))

sns.heatmap(
    pivot_top10,
    cmap="YlGnBu",
    annot=True,
    fmt=".0f"
)

plt.title("Applicants by Region (Heatmap)")

plt.xlabel("Regions", fontsize=14, labelpad=16)
plt.ylabel("KldB_code (5-digit)", fontsize=14, labelpad=16)

plt.xticks(fontsize=12)
plt.yticks(rotation=0,fontsize=12)

plt.tight_layout()
plt.savefig("../pictures/top_applicants_region_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df_bewerberboerse_2 = pd.read_csv('../data_2/output/bewerberboerse_kldb_assigned.csv')
```

```python
df_bewerberboerse_2.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df_bewerberboerse_2.shape
```

```python
df_bewerberboerse_2.drop_duplicates(subset="refnr", inplace=True)
```

```python
df_bewerberboerse_2.shape
```

```python
df_bewerberboerse_2 = df_bewerberboerse_2[df_bewerberboerse_2['similarity']>0.85]
```

```python
df_bewerberboerse_2.shape
```

```python
new_records = df_bewerberboerse_2[~df_bewerberboerse_2["refnr"].isin(df_bewerberboerse_1["refnr"])]
```

```python
new_records.shape
```

```python
old_records = df_bewerberboerse_2[df_bewerberboerse_2["refnr"].isin(df_bewerberboerse_1["refnr"])]
```

```python
old_records.shape
```

```python
removed_records = df_bewerberboerse_1[~df_bewerberboerse_1["refnr"].isin(df_bewerberboerse_2["refnr"])]
```

```python
removed_records.shape
```

```python
df_bewerberboerse_1
```

```python
freq_old = df_bewerberboerse_1["Kldb_code"].value_counts()
freq_new = df_bewerberboerse_2["Kldb_code"].value_counts()
```

```python
comparison = pd.concat([freq_old, freq_new], axis=1)
comparison.columns = ["old", "new"]
comparison = comparison.fillna(0)
```

```python
comparison["change"] = comparison["new"] - comparison["old"]
```

```python
top_increase = comparison.sort_values("change", ascending=False).head(10)

top_increase["change"].plot(kind="barh", figsize=(10,6))

plt.title("Increase in applicants")

plt.xlabel("Change in Number of applicants", fontsize=14, labelpad=16)
plt.ylabel("KldB_code (5-digit)", fontsize=14, labelpad=16)

plt.savefig("../pictures/top_increased_applicants.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
top_decrease = comparison.sort_values("change").head(10)

top_decrease["change"].plot(kind="barh", figsize=(10,6))

plt.title("Decrease in applicants")

plt.xlabel("Change in Number of applicants", fontsize=14, labelpad=16)
plt.ylabel("KldB_code (5-digit)", fontsize=14, labelpad=16)

plt.savefig("../pictures/top_decreased_applicants.png", dpi=300, bbox_inches="tight")
plt.show()
```
