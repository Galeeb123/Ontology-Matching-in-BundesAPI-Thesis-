```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Analysing jobsuche data

```python
df_jobsuche_1 = pd.read_csv('../data_1/output/jobsuche_kldb_assigned.csv')
```

```python
list(df_jobsuche_1.columns)
```

```python
df_jobsuche_1
```

```python
df_jobsuche_1.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df_jobsuche_1.drop_duplicates(subset="refnr", inplace=True)
```

```python
df_jobsuche_1.shape
```

```python
df_jobsuche_1 = df_jobsuche_1[df_jobsuche_1['similarity']>0.85]
```

```python
df_jobsuche_1.shape
```

```python
df_jobsuche_1['arbeitsort.region'].unique()
```

```python
frequency_jobs_1 = df_jobsuche_1.groupby("Kldb_code").size().reset_index(name="count")
```

```python
frequency_jobs_1
```

```python
frequency_jobs_1
```

```python
frequency_jobs_1 = frequency_jobs_1.sort_values(by="count", ascending=False)
```

```python
frequency_jobs_1
```

```python
frequency_jobs_1["Kldb_code"] = frequency_jobs_1["Kldb_code"].astype(str)
```

```python
frequency_jobs_1.info()
```

```python
top10_job_1 = frequency_jobs_1.head(10)

plt.figure(figsize=(10,6))

plt.barh(top10_job_1["Kldb_code"], top10_job_1["count"])

plt.title("Top 10 job offered")

plt.xlabel("Count", fontsize=14, labelpad=16)
plt.ylabel("KldB-2010 (5-digit)", fontsize=14, labelpad=16)

plt.gca().invert_yaxis()
plt.savefig("../pictures/top_jobs_offered.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df_jobsuche_1[df_jobsuche_1['Kldb_code']==62102]['beruf'].unique()
```

```python
df_jobsuche_1[df_jobsuche_1['Kldb_code']==51593]['similarity'].unique()
```

```python
job_region = (
    df_jobsuche_1
    .groupby(["arbeitsort.region", "Kldb_code"])
    .size()
    .reset_index(name="count")
)
```

```python
job_region
```

```python
pivot_table = job_region.pivot(
    index="Kldb_code",
    columns="arbeitsort.region",
    values="count"
).fillna(0)
```

```python
pivot_table
```

```python
top10 = pivot_table.sum(axis=1).sort_values(ascending=False).head(10)

pivot_top10 = pivot_table.loc[top10.index]
```

```python
pivot_top10
```

```python
import matplotlib.pyplot as plt

pivot_top10.plot(
    kind="bar",
    figsize=(12,7)
)

plt.title("Top jobs by Region")
plt.xlabel("Occupation")
plt.ylabel("Number of jobs Offers")

plt.xticks(rotation=45)
plt.legend(title="Region")

plt.tight_layout()
plt.show()
```

```python
top_regions = df_jobsuche_1["arbeitsort.region"].value_counts().head(5).index

pivot_filtered = pivot_top10[top_regions]

pivot_filtered.plot(kind="bar", figsize=(12,7))

plt.title("Top jobs by Region")

plt.xlabel("KldB_code (5-digit)", fontsize=14, labelpad=16)
plt.ylabel("Count", fontsize=14, labelpad=16)

plt.xticks(rotation=0)
plt.legend(title="Region")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    pivot_top10,
    cmap="YlGnBu",
    annot=True,
    fmt=".0f"
)

plt.title("Jobs by Region (Heatmap)")

plt.xlabel("Regions", fontsize=14, labelpad=16)
plt.ylabel("KldB_code (5-digit)", fontsize=14, labelpad=16)

plt.xticks(fontsize=12)
plt.yticks(rotation=0,fontsize=12)

plt.tight_layout()
plt.savefig("../pictures/top_jobs_region_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df_jobsuche_2 = pd.read_csv('../data_2/output/jobsuche_kldb_assigned.csv')
```

```python
df_jobsuche_2.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df_jobsuche_2.shape
```

```python
df_jobsuche_2.drop_duplicates(subset="refnr", inplace=True)
```

```python
df_jobsuche_2.shape
```

```python
df_jobsuche_2 = df_jobsuche_2[df_jobsuche_2['similarity']>0.85]
```

```python
df_jobsuche_2.shape
```

```python
new_records = df_jobsuche_2[~df_jobsuche_2["refnr"].isin(df_jobsuche_1["refnr"])]
```

```python
new_records.shape
```

```python
old_records = df_jobsuche_2[df_jobsuche_2["refnr"].isin(df_jobsuche_1["refnr"])]
```

```python
old_records.shape
```

```python
removed_records = df_jobsuche_1[~df_jobsuche_1["refnr"].isin(df_jobsuche_2["refnr"])]
```

```python
removed_records.shape
```

```python
df_jobsuche_1
```

```python
freq_old = df_jobsuche_1["Kldb_code"].value_counts()
freq_new = df_jobsuche_2["Kldb_code"].value_counts()
```

```python
comparison = pd.concat([freq_old, freq_new], axis=1)
comparison.columns = ["old", "new"]
comparison = comparison.fillna(0)
```

```python
comparison["change"] = comparison["new"] - comparison["old"]
```

```python
top_increase = comparison.sort_values("change", ascending=False).head(10)

top_increase["change"].plot(kind="barh", figsize=(10,6))

plt.title("Increase in Training Offers")

plt.xlabel("Change in Number of Trainings", fontsize=14, labelpad=16)
plt.ylabel("KldB_code (5-digit)", fontsize=14, labelpad=16)

plt.savefig("../pictures/top_increased_jobs.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
top_decrease = comparison.sort_values("change").head(10)

top_decrease["change"].plot(kind="barh", figsize=(10,6))

plt.title("Decrease in Training Offers")

plt.xlabel("Change in Number of Trainings", fontsize=14, labelpad=16)
plt.ylabel("KldB_code (5-digit)", fontsize=14, labelpad=16)

plt.savefig("../pictures/top_decreased_jobs.png", dpi=300, bbox_inches="tight")
plt.show()
```
